```python
import pygame
import random

# Initialize Pygame
pygame.init()

# Screen dimensions
WIDTH, HEIGHT = 800, 300

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
GRAY = (200, 200, 200)

# Initialize screen
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Dino Game")

# Clock
clock = pygame.time.Clock()

# Dino attributes
DINO_WIDTH, DINO_HEIGHT = 50, 50
dino_x, dino_y = 50, HEIGHT - DINO_HEIGHT - 10
dino_velocity = 0
GRAVITY = 1
JUMP_FORCE = -15

# Obstacle attributes
OBSTACLE_WIDTH = 20
OBSTACLE_HEIGHT = 50
obstacle_x = WIDTH
obstacle_y = HEIGHT - OBSTACLE_HEIGHT - 10
obstacle_speed = 7

# Scoring
score = 0
font = pygame.font.Font(None, 36)

def draw_dino():
    """Draw the dinosaur."""
    pygame.draw.rect(screen, BLACK, (dino_x, dino_y, DINO_WIDTH, DINO_HEIGHT))

def draw_obstacle():
    """Draw the obstacle."""
    pygame.draw.rect(screen, RED, (obstacle_x, obstacle_y, OBSTACLE_WIDTH, OBSTACLE_HEIGHT))

def display_score(score):
    """Display the current score."""
    score_text = font.render(f"Score: {score}", True, BLACK)
    screen.blit(score_text, (WIDTH - 150, 20))

def check_collision():
    """Check if the dino collides with the obstacle."""
    if (dino_x < obstacle_x + OBSTACLE_WIDTH and
        dino_x + DINO_WIDTH > obstacle_x and
        dino_y < obstacle_y + OBSTACLE_HEIGHT and
        dino_y + DINO_HEIGHT > obstacle_y):
        return True
    return False

def game_loop():
    """Main game loop."""
    global dino_y, dino_velocity, obstacle_x, score

    running = True
    while running:
        screen.fill(WHITE)

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_SPACE and dino_y == HEIGHT - DINO_HEIGHT - 10:
                    dino_velocity = JUMP_FORCE  # Make the dino jump

        # Dino movement
        dino_y += dino_velocity
        dino_velocity += GRAVITY
        if dino_y > HEIGHT - DINO_HEIGHT - 10:  # Prevent falling below ground level
            dino_y = HEIGHT - DINO_HEIGHT - 10

        # Obstacle movement
        obstacle_x -= obstacle_speed
        if obstacle_x < -OBSTACLE_WIDTH:  # Reset obstacle position
            obstacle_x = WIDTH
            score += 1

        # Check collision
        if check_collision():
            screen.fill(GRAY)
            text = font.render("Game Over! Press R to Restart", True, BLACK)
            screen.blit(text, (WIDTH // 3, HEIGHT // 2))
            pygame.display.flip()

            restart = False
            while not restart:
                for event in pygame.event.get():
                    if event.type == pygame.KEYDOWN and event.key == pygame.K_r:
                        game_loop()
                    elif event.type == pygame.QUIT:
                        running = False
                        restart = True
            break

        # Draw elements
        draw_dino()
        draw_obstacle()
        display_score(score)

        pygame.display.flip()
        clock.tick(30)

    pygame.quit()

# Run the game
game_loop()
```
