```python
import pandas as pd

# Load your data (change 'data.csv' to your file name)
data = pd.read_csv("word.csv")

# Check the first few rows
print(data.head())
```

```python
print(data['What do you wish to achieve by taking this course?'].dtype) 
```

```python
# Replace NaN with an empty string
data['What do you wish to achieve by taking this course?'] = data['What do you wish to achieve by taking this course?'].fillna("")

# Ensure all entries are strings
data['What do you wish to achieve by taking this course?'] = data['What do you wish to achieve by taking this course?'].astype(str)
```

```python
print(data['What do you wish to achieve by taking this course?'].dtype)
```

```python
from textblob import TextBlob

# Apply sentiment analysis
data['sentiment'] = data['What do you wish to achieve by taking this course?'].apply(lambda x: TextBlob(x).sentiment.polarity)

print(data[['What do you wish to achieve by taking this course?', 'sentiment']])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Combine all responses into one text
all_text = ' '.join(data['What do you wish to achieve by taking this course?'])

# Create a word cloud to visualize common words
wordcloud = WordCloud(width=1920, height=1080).generate(all_text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
data.to_csv("sentiment_analysis_results.csv", index=False)
```
